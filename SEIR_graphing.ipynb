```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#winter games
# Parameters
#gamma at 7
#with only 1 infected entering village
gamma = 7
alpha = 1/6
tau = 2.9
N = 1500
f = 0.1
dt = 0.1
T = 30
steps = int(T/dt)

# Arrays
S = np.zeros(steps)
E = np.zeros(steps)
I = np.zeros(steps)
R = np.zeros(steps)

# Initial conditions
S[0] = 1499
I[0] = 1

delay_steps = int(tau/dt)

for t in range(1, steps):
    
    if t - delay_steps >= 0:
        S_tau = S[t - delay_steps]
        I_tau = I[t - delay_steps]
    else:
        S_tau = S[0]
        I_tau = I[0]
    
    dS = -gamma * S[t-1] * I[t-1] / N
    dE = gamma * S[t-1] * I[t-1] / N - gamma * S_tau * I_tau / N
    dI = gamma * S_tau * I_tau / N - alpha * I[t-1]
    dR = alpha * f * I[t-1]
    
    S[t] = S[t-1] + dS*dt
    E[t] = E[t-1] + dE*dt
    I[t] = I[t-1] + dI*dt
    R[t] = R[t-1] + dR*dt

time = np.linspace(0, T, steps)

plt.plot(time, S, label="Susceptible")
plt.plot(time, E, label="Exposed")
plt.plot(time, I, label="Infectious")
plt.plot(time, R, label="Recovered")
plt.legend()
plt.xlabel("Time in Days")
plt.ylabel("Poplutation")
plt.show()
```
